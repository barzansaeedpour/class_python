### **جلسه ۶ – تاپل و دیکشنری**

</div>


<div dir="rtl">
تاپل‌ها (`tuple`) شبیه لیست هستند اما **غیرقابل تغییر (immutable)** می‌باشند.  
یعنی بعد از ایجاد، نمی‌توان مقادیر داخل آن را تغییر داد، حذف کرد یا عنصر جدیدی به آن افزود.  
از تاپل‌ها زمانی استفاده می‌کنیم که بخواهیم مجموعه‌ای از داده‌ها ثابت بماند.
</div>


```python
# ایجاد تاپل
my_tuple = (1, 2, 3, "سلام")
print(my_tuple)

# دسترسی به عناصر
print(my_tuple[0])  
print(my_tuple[-1])
```


<div dir="rtl">
دیکشنری (`dict`) نوعی ساختار داده است که داده‌ها را به صورت **کلید: مقدار (key:value)** ذخیره می‌کند.  
کلیدها باید یکتا باشند و معمولا از نوع رشته یا عدد می‌باشند.
</div>


```python
# ایجاد دیکشنری
student = {
    "name": "Ali",
    "age": 20,
    "major": "Computer Science"
}

print(student)

# دسترسی به مقدار یک کلید
print(student["name"])
```


<div dir="rtl">
متدهای مهم دیکشنری:
- `keys()` : لیست کلیدها را برمی‌گرداند
- `values()` : لیست مقادیر را برمی‌گرداند
- `items()` : جفت‌های کلید و مقدار را برمی‌گرداند
</div>


```python
student = {
    "name": "Sara",
    "age": 22,
    "major": "Mathematics"
}

print(student.keys())
print(student.values())
print(student.items())
```
