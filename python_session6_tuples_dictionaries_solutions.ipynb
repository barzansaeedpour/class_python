```python
# -----------------------------
# تمرین‌های ساده
# -----------------------------

# 1. یک تاپل از اعداد بسازید و مجموع آن‌ها را محاسبه کنید.
numbers = (10, 20, 30, 40, 50)   # تعریف یک تاپل از اعداد
print("مجموع اعداد:", sum(numbers))  # استفاده از تابع sum برای محاسبه مجموع
```

```python

# 2. یک تاپل شامل نام، سن و شهر خود ایجاد کنید و مقادیر آن را چاپ کنید.
person = ("بارزان", 27, "سنندج")  
print("نام:", person[0])
print("سن:", person[1])
print("شهر:", person[2])
```

```python

# 3. یک دیکشنری شامل سه رنگ (کلید: نام رنگ، مقدار: کد هگز رنگ) بسازید و آن را چاپ کنید.
colors = {
    "قرمز": "#FF0000",
    "سبز": "#00FF00",
    "آبی": "#0000FF"
}
print("دیکشنری رنگ‌ها:", colors)

```

```python

# -----------------------------
# تمرین‌های متوسط
# -----------------------------

# 4. یک دیکشنری شامل اطلاعات یک کتاب و چاپ کلید-مقدارهای آن با متد items()
book = {
    "عنوان": "صد سال تنهایی",
    "نویسنده": "گابریل گارسیا مارکز",
    "سال انتشار": 1967
}
for key, value in book.items():   # items() همه کلید-مقدارها را برمی‌گرداند
    print(f"{key}: {value}")
```

```python

# 5. یک دیکشنری شامل اطلاعات چند دانش‌آموز و محاسبه میانگین نمرات
students = {
    "علی": 18,
    "سارا": 15,
    "مریم": 19,
    "حسین": 17
}
average = sum(students.values()) / len(students)  # مجموع مقادیر / تعداد
print("میانگین نمرات:", average)
```

```python

# 6. یک تاپل شامل چند رشته و چاپ تعداد کاراکترهای هر رشته
words = ("Python", "AI", "Programming", "Data")
for w in words:
    print(f"رشته '{w}' دارای {len(w)} کاراکتر است.")


# -----------------------------
# تمرین‌های پیشرفته
# -----------------------------

```

```python

# 7. فرهنگ لغت ساده انگلیسی-فارسی
dictionary = {
    "apple": "سیب",
    "book": "کتاب",
    "car": "ماشین"
}
word = input("یک کلمه انگلیسی وارد کنید: ")
if word in dictionary:
    print("معادل فارسی:", dictionary[word])
else:
    print("کلمه مورد نظر در فرهنگ لغت وجود ندارد.")
```

```python

# 8. دیکشنری شامل کشورها و پایتخت‌ها
countries = {
    "ایران": "تهران",
    "فرانسه": "پاریس",
    "آلمان": "برلین",
    "ژاپن": "توکیو"
}
country = input("نام کشور را وارد کنید: ")
if country in countries:
    print("پایتخت:", countries[country])
else:
    print("اطلاعاتی درباره این کشور موجود نیست.")
```

```python

# 9. پیدا کردن گران‌ترین محصول در دیکشنری
products = {
    "موبایل": 25000000,
    "لپ‌تاپ": 45000000,
    "ساعت": 7000000,
    "تلویزیون": 30000000
}
# با استفاده از max می‌توان کلید محصول با بیشترین قیمت را یافت
expensive = max(products, key=products.get)
print("گران‌ترین محصول:", expensive, "با قیمت", products[expensive])
```

```python

# 10. یک تاپل دوبعدی از مختصات نقاط و محاسبه فاصله از مبدا
points = ((3, 4), (5, 12), (8, 15), (7, 24))
for x, y in points:
    distance = (x**2 + y**2) ** 0.5  # استفاده از قضیه فیثاغورس
    print(f"نقطه ({x}, {y}) فاصله‌اش از مبدا = {distance:.2f}")
```
